```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips
```

# Relplot

```python
sns.relplot(x = "total_bill",y = "tip",data = tips);
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tips,hue = "sex");
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tips,hue = "smoker");
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tips,hue = "smoker",style="sex");
```

```python
sns.relplot(x = "total_bill",y = "tip",data = tips,hue = "smoker",style="sex",size="day");
```

```python
data = pd.read_csv("Advertising.csv",index_col = 0)
data
```

```python
sns.relplot(x="radio",y="TV",data = data,hue="sales");
```
