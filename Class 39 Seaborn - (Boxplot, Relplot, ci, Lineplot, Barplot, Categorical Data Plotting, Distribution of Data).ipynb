```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

# Boxplot

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.boxplot(x='day',y='total_bill',data = tips)
```

```python
sns.boxplot(x='sex',y='tip',data = tips);
```

```python
from numpy.random import randn
```

```python
df = pd.DataFrame(dict(time = np.arange(500),value = randn(500).cumsum()))
df
```

```python
sns.relplot(x='time',y='value',data=df,kind='line');
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
fmri[['timepoint','signal','event']].max()
```

```python
sns.set(style = 'darkgrid')
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,ci=99);
```

```python
sns.relplot(x='timepoint',y='signal',hue = 'event',kind='line',data=fmri,ci=99);
```

```python
sns.relplot(x='timepoint',y='signal',hue = 'event',kind='line',data=fmri,marker=True);
```

```python
sns.relplot(x='timepoint',y='signal',hue = 'event',kind='line',data=fmri,marker=True,dashes=False);
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',col='day',data=tips);
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',col='day',col_wrap=2,data=tips);
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

# Lineplot

```python
sns.lineplot(x='year',y='passengers',data=flights,ci=False);
```

```python
flights['year'].unique()
```

```python
x=flights[flights['year']==1954]
x
```

```python
sns.barplot(x='month',y='passengers',data=x);
plt.xticks(rotation=90);
```

# Categorical Data plotting

```python
sns.catplot(x='day',y='total_bill',data=tips);
```

```python
tips.day.value_counts()
```

```python
sns.catplot(x = 'total_bill',y='day',data=tips);
```

```python
sns.catplot(x = 'smoker', y = 'tip', data = tips, order= ['Yes', 'No']);
```

```python
sns.catplot(x='day',y='total_bill',data=tips,jitter=False);
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='swarm');
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='box');
```
